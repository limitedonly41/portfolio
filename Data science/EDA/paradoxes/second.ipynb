```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
eps = 0.005
p = 0.49
q = 0.74
r = 0.09


def gameA(outcome,j):
    random.seed(j)
    if random.random() < (p):
        return outcome + 1
    else:
        return outcome - 1




def playGame(numIter, numFlip, gameF):
    npOutcome = np.zeros((numIter, numFlip))
    for i in range(numIter):
        outcome = []
        tmp = 0
        for j in range(numFlip):
            j = j*1.5
            tmp = gameF(tmp,j)
            outcome.append(tmp)
        npOutcome[i, :] = outcome
    return np.mean(npOutcome, 0)


numIter = 10
numFlip = 10000
# Game A plot

resultA = playGame(numIter, numFlip, gameA)
plt.xlabel('  ')
plt.ylabel('')
plt.title('Game A | p=0.49 ')

print(type(resultA))

my_list = list(resultA)
new_list = [x+1000 for x in my_list]
resultA = np.array(new_list)
plt.plot(resultA)
print(resultA[157])
plt.show()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
eps = 0.005
p = 0.49
q = 0.74
r = 0.09


def gameB(outcome,j):
    random.seed(j)
    if outcome % 3 == 0:
        if random.random() < (r):
            return outcome + 1
        else:
            return outcome - 1
    else:
        if random.random() < (q):
            return outcome + 1
        else:
            return outcome - 1




def playGame(numIter, numFlip, gameF):
    npOutcome = np.zeros((numIter, numFlip))
    for i in range(numIter):
        outcome = []
        tmp = 0
        for j in range(numFlip):
            j = j*3
            tmp = gameF(tmp,j)
            outcome.append(tmp)
        npOutcome[i, :] = outcome
    return np.mean(npOutcome, 0)


numIter = 10
numFlip = 10000
# Game B plot


resultB = playGame(numIter, numFlip, gameB)
plt.xlabel('  ')
plt.ylabel('')
plt.title('Game B | q=0.74 r=0.09')

my_list = list(resultB)
new_list = [x+1000 for x in my_list]
resultB = np.array(new_list)

plt.plot(resultB)
plt.show()
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
eps = 0.005
p = 0.49
q = 0.74
r = 0.09


def randomSwitchGame(outcome,j):
    random.seed(j)
    if random.random() < 0.5:
        return gameA(outcome,j)
    else:
        return gameB(outcome,j)




def playGame(numIter, numFlip, gameF):
    npOutcome = np.zeros((numIter, numFlip))
    for i in range(numIter):
        outcome = []
        tmp = 0
        for j in range(numFlip):
            j = j*10
            tmp = gameF(tmp,j)
            outcome.append(tmp)
        npOutcome[i, :] = outcome
    return np.mean(npOutcome, 0)


numIter = 10
numFlip = 10000
# Game B plot


# Game random switch plot
resultRandomSwitch = playGame(numIter, numFlip, randomSwitchGame)
plt.xlabel('  ')
plt.ylabel('')
plt.title('Game Random Switch | p=0.49 q=0.74 r=0.09')

my_list = list(resultRandomSwitch)
new_list = [x+1000 for x in my_list]
resultRandomSwitch = np.array(new_list)

plt.plot(resultRandomSwitch)
plt.show()
```
