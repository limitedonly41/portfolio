```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def ll(x):
    ind = 0
    ll_array = []
    temp_mm = 0
    for m in x:
        temp_mm = stats.norm.rvs(loc=m,scale=100, size=1, random_state = None)
        ll_array.append(temp_mm)
    return max(ll_array)
```

```python
max_likely_x = []
x = np.arange(0, 100, 1).tolist()
for n in range(0,10000,1):
    press_count = []
    for i in x:
        press_count.append(stats.norm.rvs(loc=60,scale=100, size=1, random_state = None))
        if press_count[i] < 0:
            press_count[i] = [0.0001]
    maxn = ll(press_count)
    max_likely_x.append(maxn)
```

```python
max_likely_y = []
x = np.arange(0, 100, 1).tolist()
for n in range(0,10000,1):
    press_count = []
    for i in x:
        press_count.append(stats.norm.rvs(loc=60,scale=100, size=1, random_state = None))
        if press_count[i] < 0:
            press_count[i] = [0.0001]
    maxn = ll(press_count)
    max_likely_y.append(maxn)
```

```python
max_likely_z = []
x = np.arange(0, 100, 1).tolist()
for n in range(0,10000,1):
    press_count = []
    for i in x:
        press_count.append(stats.norm.rvs(loc=60,scale=100, size=1, random_state = None))
        if press_count[i] < 0:
            press_count[i] = [0.0001]
    maxn = ll(press_count)
    max_likely_z.append(maxn)
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
tr = np.arange(0, 10000, 1).tolist()
ax.plot(tr, max_likely_x)
ax.set(xlabel='n', ylabel='Оценки',
       title='Оценки метода макс правдоподобия Y')
ax.grid()

plt.show()
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
tr = np.arange(0, 10000, 1).tolist()
ax.plot(tr, max_likely_y)
ax.set(xlabel='n', ylabel='Оценки',
       title='Оценки метода макс правдоподобия Y')
ax.grid()

plt.show()
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
tr = np.arange(0, 10000, 1).tolist()
ax.plot(tr, max_likely_z)
ax.set(xlabel='n', ylabel='Оценки',
       title='Оценки метода макс правдоподобия Y')
ax.grid()

plt.show()
```

```python
def s(x,y,z):
    return (x-60)**2+(y-70)**2+(z-80)**2
```

```python
error = []
for n in range(0,10,1):
    error.append(s(max_likely_x[n],max_likely_y[n],max_likely_z[n]))
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
tr = np.arange(0, 10000, 1).tolist()
ax.plot(tr, error)
ax.set(xlabel='n', ylabel='Ошибка',
       title='суммарная ошибка')
ax.grid()

plt.show()
```

```python
from scipy.stats.stats import pearsonr
nums = np.arange(0, 10000, 1).tolist()
graph_coeff = []
for m in nums:
    coeff = pearsonr(max_likely_x, max_likely_y)
    graph_coeff.append(coeff[0])
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
ax.plot(nums, graph_coeff)
ax.set(xlabel='N', ylabel='Значение',
       title='График выборочной корреляции X и Y')
ax.grid()

plt.show()
```
