```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
start_time = 480
end_time = 1200

time_wait = [5,10]
bus_times = []
bus_times_hour = []
time = start_time
data = []
```

```python
for i in range(0,10000):
    while time < end_time:
        passengers = choice = np.random.choice(time_wait)
        time = time + choice
        bus_times.append(time)
        hour = int(time / 60)
        bus_times_hour.append(hour)
        data.append(passengers)
    time = start_time
# так как в 20:00 автобусы не ходят, то удалим посденеее время
del bus_times[-1]
del bus_times_hour[-1]
# так как в 20:00 автобусы не ходят, то удалим последних пасссажиров
del data[-1]
```

```python
from collections import Counter
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
count = Counter(bus_times_hour)
plt.bar(count.keys(), count.values())
for x,y in zip(count.keys(),count.values()):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center', # horizontal alignment can be left, right or center
                 fontsize=8.2) 

plt.xticks(np.arange(8,20,1))
plt.show()
```
