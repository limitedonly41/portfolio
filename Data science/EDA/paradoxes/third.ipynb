```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as stats
x = np.arange(0, 10000, 1).tolist()
marks_russ= []
for i in x:
    marks_russ.append(stats.norm.rvs(loc=60,scale=10, size=1, random_state = None))

plt.figure(figsize=(15,15))
fig, ax = plt.subplots()

ax.plot(x, marks_russ)
ax.set(xlabel='n', ylabel='Оценки',
       title='График оценок 10 000 школьников по русскому')
ax.grid()

plt.show()
```

```python
import scipy.stats as stats
x = np.arange(0, 10000, 1).tolist()
marks_math= []
for i in x:
    marks_math.append(stats.norm.rvs(loc=60,scale=10, size=1, random_state = None))

plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
ax.plot(x, marks_math)
ax.set(xlabel='n', ylabel='Оценки',
       title='График оценок 10 000 школьников по математике')
ax.grid()

plt.show()
```

```python
marks_total = [x + y for x, y in zip(marks_russ, marks_math)]
```

```python
nums = np.arange(80, 160, 5).tolist()
total_students = []
total_pass = []
total_fail = []
for m in nums:
    count = 0
    for i in marks_total : 
        if i > m : 
            count = count + 1
#            total_pass.append(i)
#        else:
#            total_fail.append(i)
    total_students.append(count)
    total_fail.append(10000-count)
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
ax.plot(nums, total_students)
ax.set(xlabel='Оценка', ylabel='Количество студентов',
       title='График прошедших студентов от n')
ax.grid()

plt.show()
```

```python
import itertools
marks_math = np.vstack(marks_math)
marks_russ = np.vstack(marks_russ)
```

```python
marks_math = list(itertools.chain.from_iterable(marks_math))
```

```python
marks_russ = list(itertools.chain.from_iterable(marks_russ))
```

```python
from scipy.stats.stats import pearsonr
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as stats
graph_coeff = []
for m in nums:
    x = []
    y = []
    count = 0
    k = 0
    for i in marks_total :
        if k == 100000:
            break
        else:
            if i > m : 
                x.append(marks_math[k])
                y.append(marks_russ[k])
    #            total_pass.append(i)
    #        else:
    #            total_fail.append(i)
        k =k + 1

    coeff = pearsonr(x, y)
    graph_coeff.append(coeff[0])
print(graph_coeff)
```

```python
plt.figure(figsize=(15,15))
fig, ax = plt.subplots()
ax.plot(nums, graph_coeff)
ax.set(xlabel='N', ylabel='Значение',
       title='График выборочной корреляции среди прошедших в зависимости от n')
ax.grid()

plt.show()
```
